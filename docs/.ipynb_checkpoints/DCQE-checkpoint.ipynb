# Delayed-choice quantum eraser

## Introduction
In the 1999 paper by Yoon-Ho Kim, R. Yu, S.P. Kulik, Y.H. Shih and Marlan O. Scully. "[A Delayed Choice Quantum Eraser](https://arxiv.org/abs/quant-ph/9903047v1)", the authors investigate the possibility of simultaneously observing both particle-like and wave-like behavior of a photon via quantum entanglement by experiment. The experimental results show that the which-path of a quantum can be erased or marked by its entangled twin even after the registration of the quantum. They further gave the theoretical explanation for the results. The theoretical framework includes the creation operators and many more sophisticate mathematical treatments.

## Experiment setup and results (From the [paper](https://arxiv.org/abs/quant-ph/9903047v1) and [Wikipedia](https://en.wikipedia.org/wiki/Delayed-choice_quantum_eraser)) <a id='setup_results'></a>
### Setup <a id='setup'></a>
I will first briefly talk about the setup and the results obtained in the original [paper](https://arxiv.org/abs/quant-ph/9903047v1). Afterwards, I will present the theoretical model that I mentioned above.

Figure below shows a schematic diagram of the setup. Spontaneous parametric down conversion (SPDC) (for more information about SPDC, see the Wikipedia [page](https://en.wikipedia.org/wiki/Spontaneous_parametric_down-conversion)) is used to prepare the entangled two-photon state in the experiment. 
<img src="images/dcQE.png" height="317" width="450" align="center" alt="delayed choice quantum eraser" title="delayed choice Quantum Eraser"/>
<div align="center">Fig.1 <a id='figure1'></a>Experimental setup (adapted from <a href="https://en.wikipedia.org/wiki/Delayed-choice_quantum_eraser">Wikipedia</a>)</div>

An argon laser generates individual 351.1 nm photons that pass through a double-slit apparatus. In the illustration, the photon paths are color-coded as red or light blue lines to indicate which slit the photon came through.

So far, the experiment is like a conventional two-slit experiment. However, after the slits, spontaneous parametric down-conversion (SPDC) is used to prepare an entangled two-photon state. This is done by a nonlinear optical crystal BBO (beta barium borate) that converts the photon (from either slit) into two identical, orthogonally polarized entangled photons with 1/2 the frequency of the original photon. The paths followed by these orthogonally polarized photons are caused to diverge by the Glan-Thompson prism.

One of these 702.2 nm photons, referred to as the `signal` photon continues to the target detector called $D_0$. During an experiment, detector $D_0$ is scanned along its x axis, its motions controlled by a step motor. A plot of signal photon counts detected by $D_0$ versus $x$ can be examined to discover whether the cumulative signal forms an interference pattern.

The other entangled photon, referred to as the `idler` photon, is deflected by prism $PS$ that sends it along divergent paths depending on whether it came from slit A or slit B.

After the prism $PS$, the idler photons encounter beam splitters $BS_a$, $BS_b$, and $BS_c$ that each have a 50% chance of allowing the idler photon to pass through and a 50% chance of causing it to be reflected. $M_a$ and $M_b$ are mirrors.

The beam splitters and mirrors direct the idler photons towards detectors labeled $D_1$, $D_2$, $D_3$ and $D_4$. Note that:

* If an idler photon is recorded at detector $D_3$, it can only have come from slit B.
* If an idler photon is recorded at detector $D_4$, it can only have come from slit A.
* If an idler photon is detected at detector $D_1$ or $D_2$, it might have come from slit A or slit B.

The optical path length measured from slit to $D_1$, $D_2$, $D_3$, and $D_4$ is at least 2.5 m longer than the optical path length from slit to $D_0$. This means that any information that one can learn from an idler photon must be approximately 8 ns later than what one can learn from its entangled signal photon.

Detection of the idler photon by $D_3$ or $D_4$ provides delayed "which-path information" indicating whether the signal photon with which it is entangled had gone through slit A or B. On the other hand, detection of the idler photon by $D_1$ or $D_2$ provides a delayed indication that such information is **not** available for its entangled signal photon.

By using a coincidence counter, the experimenters were able to isolate the entangled signal from photo-noise, recording only events where both signal and idler photons were detected (after compensating for the 8 ns delay).

## Significance <a id='significance'></a>
This result is similar to that of the double-slit experiment, since interference is observed when it is not known from which slit the photon originates, while no interference is observed when the path is known.

However, what makes this experiment possibly astonishing is that, unlike in the classic double-slit experiment, **the choice** of whether to preserve or erase the which-path information of the `idler` **was not made until 8 ns after the position of the `signal` photon had already been measured by $D_0$**.

In other words, even though an `idler` photon is not observed until long after its entangled `signal` photon arrives at $D_0$ due to the shorter optical path for the latter, interference at $D_0$ is determined by whether a signal photon's entangled idler photon is detected at a detector that preserves its which-path information ($D_3$ or $D_4$), or at a detector that erases its which-path information ($D_1$ or $D_2$). This seems to suggest retrocausality.

However, the $\pi$ **phase shift** (mentioned in the caption of [Figure 3](#figure3)) between the interference pattern obtained in $R_{01}$ and $R_{02}$ causes the interference pattern to be **indecipherable**, as the dark fringes of one overlap with the bright fringes of the other. Therefore one would only see a bright central band. The patterns can only be "pulled" out after the detection has been made by filtering with the help of the coincidence counter. Therefore, no information can be sent back in time using this setup.

**Note 1**:
The $\pi$ phase shift is caused by the transmission-reflection unitary transformation of the beamsplitter $BS_c$ according to the author of the paper. 

**Note 2**:
$\newcommand{\ket}[1]{\left|{#1}\right\rangle}$
$\newcommand{\bra}[1]{\left\langle{#1}\right|}$
$\newcommand{\braket}[1]{\left\langle{#1}\right\rangle}$<br/>
The authors used quantum state $\ket{\Psi}$ to describe the state of the radiation field, and:

\begin{gather}
    \Psi(t_0,t_j)=\braket{0\left|E_j^{(+)}E_0^{(+)}\right|\Psi}
\end{gather}

is called the joint count probability amplitude. It is also called as the "wave function" by the authors for short. However, it is not the (spatial) wave function that we are familiar with, because here it is a function of time and not a spatial function. It is not the spatial wave function that describe the spatial distribution of the photon. For more detailed theoretical framework, please read the original paper.

## Theoretical framework (Modified and simplified) <a id='theo_framework'></a>
As mentioned above, the main goal of this project is to lay out a simpler theoretical framework or calculations to explain the results. 

### Modification and simplification made <a id='modification'></a>
The main modification to the theoretical framework that will be used later here is instead of describing the state of the radiation field using $\ket{\psi}$, we will construct of a spatial wavefunction for the photon(s) $\Psi(\mathbf{r})$ using the optical intensity distribution $I(\mathbf{r})$. $|\Psi(\mathbf{r})|^2$ is the probability distribution that the photon will be detected at certain location.

This is based on the reasonable assumption that the probability that the photon is detected at a certain location is proportional to the intensity of the optical wave. This assumption is further supported by the book "[Fundamentals of photonics](https://onlinelibrary.wiley.com/doi/book/10.1002/0471213748)", written by Saleh and Teich. (See chapter 12 section C of the book). In short:

\begin{gather}
    |\braket{\mathbf{r}|\gamma}|^2=|\Psi(\mathbf{r})|^2\propto I(\mathbf{r})\propto |U(\mathbf{r})|^2
\end{gather}

where $\ket{\gamma}$ is the state of the photon, $\braket{\mathbf{r}|\gamma}$ is the position space representation of the state of the photon, which is just $\Psi(\textbf{r})$, the quantum mechanical spatial wavefunction of the photon. $I(\textbf{r})$ is the optical intensity distribution, and $U(\textbf{r})$ is the spatial part of complex optical wavefunction.

It is not common to describe relativistic particle, such as photon, by assigning a state to it and/or write down a wave function for it. People normally describe the radiation field instead of the photon itself. However, it turns out that it will give the same prediction here (as far as I know). This is partly because here the Schrödinger equation (which is not a relativistic equation) is **not** used. We will only use the basic quantum mechanical interpretation of the wave function. I expect that some of the readers might not be convinced about this. If that's the case, one can easily replace the photon by some non-relativistic fermionic particles and replace every optical instruments in the experiment with somethings that have similar function. An example can be found in this [website](https://www.preposterousuniverse.com/blog/2019/09/21/the-notorious-delayed-choice-quantum-eraser/) from Sean Carroll.

### Detailed calculations <a id='detail'></a>
#### 1. Before Spontaneous-parametric-down (SPD) conversion <a id='b4_SPD'></a>
For simplicity, instead of using two slits, we will use two pin holes to produce nearly spherical waves, which can be approximated by paraboloidal waves at distance sufficiently far away from the holes. And we also consider monochromatic light source, with $\hbar \omega$ as the energy of the photons.

<img src="images/double-pinhole.png" height="200" width="233" align="center" alt='two pin holes' title='two pin holes'/>
<div align="center">Fig.5 Two pinholes instead of slits to produce nearly spherical waves. (adapted from <a href="https://onlinelibrary.wiley.com/doi/book/10.1002/0471213748">book</a>)</div>

Position of pin holes:
\begin{gather}
    \text{Left : }\textbf{r}_L=(-a,0,0)\hspace{0.5in}\text{Right : }\textbf{r}_R=(a,0,0)
\end{gather}

State of photon after passing through the pin hole:

\begin{equation}
    \ket{\gamma}=\frac{1}{\sqrt{2}}\ket{L_1}+\frac{1}{\sqrt{2}}\ket{R_1}
\end{equation}

where $\ket{L_1}$ denotes the state that the photon passes through the pin-hole on the left, whereas $\ket{R_1}$ denotes the state that the photon passes through the pin-hole on the right. And they are orthogonal to each other, i.e. $\braket{L_1|R_1}=0$. The factor $\left|\frac{1}{\sqrt{2}}\right|^2$ represent the 50% chance that the photon passes through one of the holes.

The position-space representation of photon state, i.e. its spatial (and temporal) wavefunction is:

\begin{equation}
    \braket{\textbf{r},t|\gamma}=\frac{1}{\sqrt{2}}\braket{\textbf{r},t|L_1}+\frac{1}{\sqrt{2}}\braket{\textbf{r},t|R_1}=\Psi(\textbf{r},t)
\end{equation}

Applying the [assumption](#modification) that I made earlier, i.e. 

\begin{gather}
    |\braket{\textbf{r}|\gamma}|^2=|\Psi(\textbf{r})|^2\propto I(\textbf{r})\propto |U(\textbf{r})|^2
\end{gather}

Let us consider a concrete example. In the paraboloidal wave approximation \[[1](#footnote1)\], where the distance from the holes is large enough, we have:

\begin{gather}
\begin{split}
    \braket{\textbf{r},t|L_1}&\propto U_{L_1}(\textbf{r})\exp(i\omega t)=\frac{A_0}{z}\exp(-ikz)\exp\left(-ik\frac{x_L^2+y^2}{2z}\right)\exp(i\omega t)\\
    \braket{\textbf{r},t|R_1}&\propto U_{R_1}(\textbf{r})\exp(i\omega t)=\frac{A_0}{z}\exp(-ikz)\exp\left(-ik\frac{x_R^2+y^2}{2z}\right)\exp(i\omega t)
\end{split}\\
    x_L=x+a\hspace{0.5in}x_R=x-a
\end{gather}

where $A_0$ is just a factor \[[2](#footnote2)\] and $\exp\left(-ik\frac{x_L^2+y^2}{2z}\right)$ is the part responsible for the paraboloidal wavefront. For simplicity, we can choose the value of $A_0$ such that the absolute square of the wave is normalized and equal-sign instead of proportional-sign can be used: 

\begin{gather}
\begin{split}
    \Psi_{L_1}(\textbf{r},t)&=\braket{\textbf{r},t|L_1}=\frac{A_0}{z}\exp(-ikz)\exp\left(-ik\frac{x_L^2+y^2}{2z}\right)\exp(i\omega t)\\
    &=\Psi_{L_1}(\textbf{r})\exp(i\omega t)
\end{split}
\end{gather}

and similar for $\Psi_{L_2}(\textbf{r},t)=\braket{\textbf{r},t|L_2}$. 

[1]<a id='footnote1'></a>: See the Fundamentals of Photonics by Saleh and Teich, subsection 2.2 Monochromatic waves: Elementary waves: Fresnel approximation of the spherical wave: The paraboloidal wave (page 46).

[2]<a id='footnote2'></a>: $A_0$ can also be a function itself, for example a gaussian shape function in $x$ and $y$.

#### 2. After SPD conversion <a id='aft_SPD'></a>
After the photons pass through the BBO crystal, some of them undergo spontaneous parametric down (SPD) conversion, and turned into two daughter photons: the `signal` photon and the `idler` photon. The `idler` photon will head towards the which-way-detector, whereas the `signal` photon will head towards the screen. The state of `idler` daughter photon can be represented as:

\begin{gather}
\begin{split}
    \ket{\gamma_i}&=\frac{1}{\sqrt{2}}\ket{L_2}\otimes\ket{A}+\frac{1}{\sqrt{2}}\ket{R_2}\otimes\ket{B}\\
    &=\frac{1}{\sqrt{2}}\ket{L_2,A}+\frac{1}{\sqrt{2}}\ket{R_2,B}
\end{split}
\end{gather}

where $\ket{A}$ and $\ket{B}$ represent photon being detected by detector $A$ or $B$ ($A$ is equivalent to detector $D_4$ and $B$ to $D_3$ of the original setup). 

<img src="images/idler_photon-equi.png" height="233" width="450" align="center" alt='detectors and eraser' title='detectors and eraser'/>
<div align="center">Fig.6<a id='figure6'></a> "Which-way" detectors: $A$ and $B$. "Eraser" part: $C$ and $D$. This is just part of the setup of the <a href='figure1'>original experiment</a> by renaming $D_3$ to $B$, $D_4$ to $A$, $D_2$ to $C$, and $D_1$ to $D$.</div>

The tensor product between $\ket{L_2}$ and $\ket{A}$; $\ket{R_2}$ and $\ket{B}$ expresses mathematically the entanglement between the path (or origin) of the photon and the detection (i.e. detection will be made by which detector).

The state of `signal` daughter photon is similar to the state of the `idler` photon, with just $\ket{L_2}$ and $\ket{R_2}$ replaced by $\ket{L_3}$ and $\ket{R_3}$, where the state of the signal and idler photon are related in a way such that conservation laws such as conservation of momentum and energy are obeyed. This relation is also known as entanglement and it is another entanglement in this system. The state of the `signal` photon can be represented by:

\begin{gather}
\begin{split}
    \ket{\gamma_s}&=\frac{1}{\sqrt{2}}\ket{L_3}\otimes\ket{A}+\frac{1}{\sqrt{2}}\ket{R_3}\otimes\ket{B}\\
    &=\frac{1}{\sqrt{2}}\ket{L_3,A}+\frac{1}{\sqrt{2}}\ket{R_3,B}
\end{split}
\end{gather}

The direction of the daughter photons are different from the parent photon. The change in direction can be expressed as rotation of coordinate system. 

<img src="images/rotation.png" height="183" width="217" align="center" alt='rotation of coordinate system' title='rotation of coordinate system'/>
<div align="center">Fig.7 Rotation of coordinate system.</div>

The frequency (wavelength) associated with the photon is changed as well after the SPD conversion, obeying conservation of energy:

\begin{gather}
    \Psi_{L_2}(\textbf{r},t)=\braket{\textbf{r},t|L_2}=\frac{A_0}{z_L'}\exp(-ik_2z_L')\exp\left(-ik_2\frac{x_L'^2+y_L'^2}{2z_L'}\right)\exp(i\omega_2 t)\\
    \hbar\omega_2=\frac{1}{2}\hbar\omega\hspace{0.25in}
    k_2=\frac{2\pi}{\lambda_2}=\frac{2\pi}{2\lambda}=\frac{1}{2}k
\end{gather}

where $\textbf{r}'=(x_L',y_L',z_L')$ is the position coordinate in the rotated coordinate system. To transform back to the "old" coordinate system, we can use the transformation formulae:

\begin{gather}
\begin{split}
    z&=z_L'\cos{\theta_L}\\
    x&=x_L'\sin{\theta_L}\cos{\varphi_L}\\
    y&=y_L'\sin{\theta_L}\sin{\varphi_L}
\end{split}
\end{gather}

where $\theta_L$ is the polar angle and $\varphi_L$ is the azimuthal angle between the new and old system. Similar for $\braket{\textbf{r},t|R_2}$ with $x=x_R'\sin{\theta_R}\cos{\varphi_R}$. Whereas for the wavefunction of the `signal` daughter photon:

\begin{gather}
    \Psi_{L_3}(\textbf{r},t)=\braket{\textbf{r},t|L_3}=\frac{A_0}{z_L''}\exp(-ik_2z_L'')\exp\left(-ik_2\frac{x_L''^2+y_L''^2}{2z_L''}\right)\exp(i\omega_2 t)
\end{gather}

where $\textbf{r}''=(x_L'',y_L'',z_L'')$ is the position coordinate in another rotated coordinate system. The transformation formulae are:

\begin{gather}
\begin{split}
    z&=z_L''\cos(-\theta_L)=z_L''\cos(\theta_L)\\
    x&=x_L''\sin(-\theta_L)\cos(-\varphi_L)=-x_L''\sin(\theta_L)\cos(\varphi_L)\\
    y&=y_L''\sin(-\theta_L)\sin(-\varphi_L)=y_L''\sin(\theta_L)\sin(\varphi_L)
\end{split}
\end{gather}

where the minus signs of the angles represent the entanglement (i.e. conservation of momentum) between the signal and idle photon. Similar for $\braket{\textbf{r},t|R_3}$ with $x=x_R''\sin(-\theta_R)\cos(-\varphi_R)$.

#### 3. Obtaining the "Which-way" information <a id='which_way'></a>

When the idler photon is detected by detector $A$, $\ket{\gamma_i}$ and $\ket{\gamma_s}$ "collapse" \[[3](#footnote3)\] into $\ket{L_2}$ and $\ket{L_3}$ respectively. For example: 

\begin{gather}
\begin{split}
    \braket{A|\gamma_s}&=\frac{1}{\sqrt{2}}\ket{L_3}\otimes\braket{A|A}+\frac{1}{\sqrt{2}}\ket{R_3}\otimes\braket{A|B}\\
    &=\frac{1}{\sqrt{2}}\ket{L_3}
\end{split}
\end{gather}

where we have used $\braket{A|B}=0$. The wavefunction is then:

\begin{gather}
\begin{split}
    \Psi_A(\textbf{r})&=\braket{\textbf{r},A|\gamma_s}\\
    &=\frac{1}{\sqrt{2}}\braket{\textbf{r}|L_3}\\
    &=\frac{A_0}{\sqrt{2}z_L''}\exp(-ik_2z_L'')\exp\left(-ik_2\frac{x_L''^2+y_L''^2}{2z_L''}\right)
\end{split}
\end{gather}

Using $|e^{i\Phi}|=1$, the probability distribution $\rho_A(\textbf{r})$ is therefore:

\begin{gather}
    \rho_A(\textbf{r})=|\Psi_A(\textbf{r})|^2=\frac{|A_0|^2}{2z_L''^2}=\text{constant on the plane }z_L=d\text{, if }|A_0|^2\text{ is constant.}
\end{gather}

Similarly, when the idler photon is detected by detector $B$, $\ket{\gamma_i}$ and $\ket{\gamma_s}$ "collapse" into $\ket{R_2}$ and $\ket{R_3}$ respectively. For example:

\begin{gather}
\begin{split}
    \braket{B|\gamma_s}&=\frac{1}{\sqrt{2}}\ket{L_3}\otimes\braket{B|A}+\frac{1}{\sqrt{2}}\ket{R_3}\otimes\braket{B|B}\\
    &=\frac{1}{\sqrt{2}}\ket{R_3}
\end{split}
\end{gather}

where we have again used $\braket{A|B}=0$.The wavefunction is then:

\begin{gather}
\begin{split}
    \Psi_B(\textbf{r})&=\braket{\textbf{r},B|\gamma_s}\\
    &=\frac{1}{\sqrt{2}}\braket{\textbf{r}|R_3}\\
    &=\frac{A_0}{\sqrt{2}z_R''}\exp(-ik_2z_R'')\exp\left(-ik_2\frac{x_R''^2+y_R''^2}{2z_R''}\right)
\end{split}
\end{gather}

Using $|e^{i\Phi}|=1$, the probability distribution $\rho_B(\textbf{r})$ is therefore:

\begin{gather}
    \rho_B(\textbf{r})=|\Psi_B(\textbf{r})|^2=\frac{|A_0|^2}{2z_R''^2}=\text{constant on the plane }z_R=d\text{, if }|A_0|^2\text{ is constant.}
\end{gather}

However, which path is taken is purely probabilistic or random (at least according to current theory). In our case, both paths are equally probable. So the total probability distribution should be the sum of the two distributions $\rho_{tot}=\rho_A+\rho_B$.

[3]<a id='footnote3'></a>: The word `collapse` is more relevant in the orthodox Copenhagen interpretation of quantum mechanics, but not in the Many-world interpretation.

## Simulation of the results when "which-way" information is obtained <a id='simu_path_known'></a>

I assume that Fraunhofer diffraction happens at the pin hole with diameter $D$ such that the optical intensity distribution takes the form \[[4](#footnote4)\]:

\begin{gather}
I(x,y)=I_0\left[\frac{2J_1(\pi D\rho/\lambda d)}{\pi D\rho/\lambda d}\right]^2\hspace{0.25in}\rho=\sqrt{x^2+y^2}
\end{gather}

where $I_0$ is the peak intensity, $J_1(.)$ is the Bessel function of order 1, $d$ is the distance between the BBO crystal and the screen, and $\lambda$ is the wavelength assign to the daughter photon. Therefore:

\begin{gather}
|A_0(x,y)|^2=I_0\left[\frac{2J_1(\pi D\rho/\lambda d)}{\pi D\rho/\lambda}\right]^2\hspace{0.25in}\rho=\sqrt{x^2+y^2}
\end{gather}

[4]<a id='footnote4'></a>: See page 124 of Fundamental of photonics, Saleh and Teich.

```python
import random
import math as m
import numpy as np
import matplotlib.pyplot as plt
import scipy.interpolate as interpolate

from scipy.integrate import simps
from scipy import special as sp

from time import sleep,time
from threading import Thread
from ipycanvas import MultiCanvas, hold_canvas,Canvas
```

### Fraunhofer diffraction code cell <a id='fraunhofer'></a>

```python
# Intensity distribution

def fraunhofer(rhodummy,d):
    D = 1e-6 # pin hole's diameter
    I0= 1
    lam = 700e-9 # wavelength lambda
    
    argdummy = np.pi*D*rhodummy/(lam*d)
    dendummy = np.pi*D*rhodummy/(lam*d)
    return(I0*(2*sp.jv(1, argdummy)/dendummy)**2)

points = 400
x = np.linspace(-1, 1, points)
y = np.linspace(-1, 1, points)

a = float(input("Please enter the separation between the pinholes, for example: 1e-2. "))

distance = float(input("Please enter the separation between crystal and screen. For example: If d = 1m, enter 1: "))

# From the pin hole on the left
xL = x - a
yL = y

# From the pin hole on the right
xR = x + a
yR = y


yesno3d = input("Do you want to see the 3D plot of the intensity? [y/n]: ")
if yesno3d == 'y':
    print("Please wait for a while...")
    
    from mpl_toolkits.mplot3d import Axes3D
    from matplotlib import cm
    
    xmesh, ymesh = np.meshgrid(x, y)
    rho = np.sqrt(xmesh**2 + ymesh**2)
    I = fraunhofer(rho, distance)
    
    xLmesh, yLmesh = np.meshgrid(xL, yL)
    xRmesh, yRmesh = np.meshgrid(xR, yR)
    
    fig = plt.figure()
    ax = Axes3D(fig)
    ax.plot_surface(xLmesh, yLmesh, I, rstride=1, cstride=1, cmap=cm.jet)
    ax.plot_surface(xRmesh, yRmesh, I, rstride=1, cstride=1, cmap=cm.jet)
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_zlabel('Intensity')
    plt.show()
else:
    print("Please proceed to the next cell.")
```

```python
# Dependence on x of the intensity

I_x = fraunhofer(x, distance) 
integral_x = simps(I_x, x)
normal_x = I_x/integral_x

# normal 2-dim plot
plt.plot(xL, normal_x, label = 'normed intensity_xL')
plt.plot(xR, normal_x, label = 'normed intensity_xR')
plt.xlabel('x')
plt.ylabel('$|\psi|^2$')
plt.legend()
plt.show()
```

```python
# Dependence on y of the intensity

I_y = fraunhofer(y, distance)
integral_y = simps(I_y, y)
normal_y = I_y/integral_y

# normal 2-dim plot
plt.plot(y, I_y)
plt.plot(y, normal_y, label = 'normed intensity_y')
plt.xlabel('y')
plt.ylabel('$|\psi|^2$')
plt.legend()
plt.show()
```

### Inverse transform sampling <a id='inv_trn_smp'></a>

To create an array of n random samples from the intensity distribution obtained when we assume Fraunhofer diffraction. 

As far as I know, the distributions that are available in `numpy` are limited to some common distributions, e.g. gaussian and uniform distribution. In order to get random numbers from other distributions that we want, we have to do inverse transform sampling. 

```python
# Create a function for inverse transform sampling

def inverse_transform_sampling(func_val, x, n=100):
    """create an array of n random samples from an arbitrary normalized distribution"""
    
    # to create the bins
    bins = np.diff(x)
    
    # find the values of the cummulative distribution function of our distribution function
    cum_values = np.cumsum(func_val[1:]*bins)
    
    # to make sure our cummulative distribution function starts from zero
    cum_values = np.insert(cum_values, 0, 0)
    
    # create a function called inv_cdf that interpolate x as a function of cum_values
    inv_cdf = interpolate.interp1d(cum_values, x) 
    
    # create an array with n elements which take their value from [0,1], which maps to our cummulative distribution fct.
    r = np.random.rand(n) 
    
    # find the x for each r
    return inv_cdf(r)

duration = int(input("Please enter the number of time steps (natural number). Example: 1000: "))
N = int(m.ceil(duration/((distance)**2))) # Number of photons drops with the square of distance btwn crystal and screen.

# create an array of N random samples from normal_x with xL as the bins
samples_xL = inverse_transform_sampling(normal_x, xL, N)

plt.hist(samples_xL, bins = x)
plt.xlabel('x')
plt.ylabel('count_L')
plt.show()

# create an array of N random samples from normal_x with xR as the bins
samples_xR = inverse_transform_sampling(normal_x, xR, N)

plt.hist(samples_xR, bins = x)
plt.xlabel('x')
plt.ylabel('count_R')
plt.show()

# create an array of N random samples from normal_y with y as the bins
samples_y = inverse_transform_sampling(normal_y, y, N)

plt.hist(samples_y, bins = y)
plt.xlabel('y')
plt.ylabel('count')
plt.show()
```

### Simulation <a id='sim_no_interference'></a>

To simulate the pattern formed on the screen using the generated random numbers drawn from our distributions.

```python
canvas = Canvas(width = 300, height = 300)
display(canvas)

xscale = 100 # To zoom in (or out) in the x-direction
yscale = 100 # To zoom in (or out) in the y-direction
def draw():
    for i in range(len(samples_xL)):
        canvas.global_alpha = 0.5
        radius=1
        randomizer = random.uniform(0, 1) # To simulate the random or probabilistic nature of which path is taken.
        if randomizer<0.5:
            canvas.fill_style='red'
            canvas.fill_arc(samples_xL[i]*xscale + 150, samples_y[i]*yscale + 150, radius, 0, 2*np.pi)
        else:
            canvas.fill_style='blue'
            canvas.fill_arc(samples_xR[i]*xscale + 150, samples_y[i]*yscale + 150, radius, 0, 2*np.pi)
        sleep(0.03)
        
simulation = Thread(target = draw)

simulation.start()
```

#### 4. Erase the "Which-way" information <a id='erase'></a>
The "which-way" information is erased when the photon is detected by detector $C$ or $D$. 

<img src="images/idler_photon-equi.png" height="233" width="450" align="center" alt='detectors and eraser' title='detectors and eraser'/>
<div align="center">Fig.8  A copy of <a href='figure6'>Figure 6</a>.</div>

The states of the signal photon when detected by detector $C$ and $D$ are:

\begin{gather}
\begin{split}
    \ket{C}&=\frac{1}{\sqrt{2}}\ket{L_2,A}+\frac{1}{\sqrt{2}}\ket{R_2,B}\\
    \ket{D}&=\frac{1}{\sqrt{2}}\ket{L_2,A}-\frac{1}{\sqrt{2}}\ket{R_2,B}
\end{split}
\end{gather}

where $\ket{L_2,A}=\ket{L_2}\otimes\ket{A}$, and $\ket{C}$ is essentially the same as $\ket{\gamma_i}$. The physical explanation for the difference in sign is the transmission reflection unitary transformation of the beamsplitter $BS_c$, as suggested by the authors of the original paper. (Actually only some types of beam splitters will cause this phase shift. See a website from [RP photonics](https://www.rp-photonics.com/beam_splitters.html) and the section `Phase shift` on this [Wikipedia](https://en.wikipedia.org/wiki/Beam_splitter) page.)

Let us check whether $\ket{C}$ and $\ket{D}$ have the correct expressions by checking whether the detection of the signal photon by detector $C$ and $D$ tells us that the probability that the photon came from the left or right pin hole is $50\%$ for both cases. Using $\braket{L_2|R_2}=0$ and $\braket{A|B}=0$, we then have:

\begin{gather}
\begin{split}
    |\braket{L_2,A|C}|^2&=\frac{1}{2}=|\braket{R_2,B|C}|^2\\
    |\braket{L_2,A|D}|^2&=\frac{1}{2}=|\braket{R_2,B|D}|^2
\end{split}
\end{gather}

as expected. The wavefunction of the signal photon when detected at $C$:

\begin{equation}
    \Psi_C(\textbf{r})=\braket{\textbf{r}|C}=\braket{\textbf{r}|\gamma_i}=\frac{1}{\sqrt{2}}\left(\braket{\textbf{r}|L_2,A}+\braket{\textbf{r}|R_2,B}\right)
\end{equation}

Whereas the wavefunction of the signal photon when detected at $D$:

\begin{equation}
    \Psi_D(\textbf{r})=\braket{\textbf{r}|D}=\frac{1}{\sqrt{2}}(\braket{\textbf{r}|L_2,A}-\braket{\textbf{r}|R_2,B})
\end{equation}

#### 5. Interference pattern and phase shift <a id='interference'></a>
Now we would like to compare the interference pattern on the screen when the signal photon is detected at $C$ with the one when detected at $D$. The interference equation (Equation 2.5.4 on page 58 of Fundamental of Photonics):

\begin{gather}
    I=I_1+I_2+2\sqrt{I_1I_2}\cos{\Phi}
\end{gather}

where $\Phi=\Phi_2-\Phi_1$ is the phase difference between the interfering waves . 


Now let us check how the intensity is distributed for: 

\begin{gather}
    \Psi_I(\textbf{r})=\braket{\textbf{r}|\gamma}=U_I(\textbf{r})=\frac{1}{\sqrt{2}}\braket{\textbf{r}|L_1}+\frac{1}{\sqrt{2}}\braket{\textbf{r}|R_1}=\frac{1}{\sqrt{2}}(\Psi_L+\Psi_R)
\end{gather}

$\ket{L_1}$ and $\ket{R_1}$ are considered here instead of $\ket{L_2}$, $\ket{R_2}$ or $\ket{L_3}$, $\ket{R_3}$ for simplicity of the notation. However the results are the same. Let us assume that the separation from the source to the screen is $d$, so $z=d$. Since things are symmetric in $y$-direction, for simplicity, we consider the case when $y=0$. This yields: 

\begin{gather}
\begin{split}
    \Psi_L(\textbf{r})&=\frac{A_0}{d}\exp(-ikd)\exp\left(-ik\frac{x_L^2}{2d}\right)\\
    &=\frac{A_0}{d}\exp\left[-ik\left(d+\frac{x_L^2}{2d}\right)\right]=\frac{A_0}{d}e^{i\Phi_L}\\
    \Psi_R(\textbf{r})&=\frac{A_0}{d}\exp\left[-ik\left(d+\frac{x_R^2}{2d}\right)\right]=\frac{A_0}{d}e^{i\Phi_R}
\end{split}
\end{gather}

And the phase difference between $\Psi_L(\textbf{r})$ and $\Psi_R(\textbf{r})$ is:

\begin{gather}
\begin{split}
    \Phi_I&=\Phi_R-\Phi_L=\left[k\left(d+\frac{x_L^2}{2d}\right)\right]-\left[k\left(d+\frac{x_R^2}{2d}\right)\right]\\
    &=k\left(\frac{x_L^2-x_R^2}{2d}\right)=k\left(\frac{(x-a)^2-(x+a)^2}{2d}\right)\\
    &=k\left(\frac{-4xa}{2d}\right)=-k\left(\frac{2xa}{d}\right)
\end{split}
\end{gather}

Now we can compute the intensity distribution using the interference equation (The constant prefactor is neglected for simplicity):

\begin{gather}
\begin{split}
    I_I(x)&=I_L+I_R+2\sqrt{I_LI_R}\cos\left[-k\left(\frac{2xa}{d}\right)\right]\\
    &=2I_0+2I_0\cos\left[k\left(\frac{2xa}{d}\right)\right]\\
    &=2I_0\left(1+\cos\left[k\left(\frac{2xa}{d}\right)\right]\right)
\end{split}
\end{gather}

where we have used: 

\begin{gather*}
    |e^{i\Phi}|=1\\
    \cos(-\Phi)=\cos(\Phi)\\
    I_L=|\Psi_L(x)|^2=I_0=I_R=|\Psi_R(x)|^2
\end{gather*}

### Interference code cells <a id='itf_codes'></a>

```python
def psi(A, d, x):
    """To create the wavefunction"""
    i = 1j
    lam = 700e-9 # wavelength lambda
    k = 2*np.pi/lam # wavenumber k
    xdummy = d + ((x**2)/(2*d))
    return(A*np.exp(-i*k*xdummy))

# It might cause some confusion here because of the sign. The sign is so because the origin of our coordinate system is
# at the middle between the 2 pinholes.
xLinterfere = x + a
xRinterfere = x - a
PsiL = psi(np.sqrt(I_x), distance, xLinterfere)
PsiR = psi(np.sqrt(I_x), distance, xRinterfere)
Psi = psi(np.sqrt(I_x), distance, x) # Just for comparison purpose

Psi1 = PsiL + PsiR

IntensL = np.absolute(PsiL)**2
IntensR = np.absolute(PsiR)**2
Intens = np.absolute(Psi)**2
IntensPsi1 = np.absolute(Psi1)**2

# To create values for the function we obtained above using the interference equation
# which is then used to compared with the values we obtained by directly taking the absolute square of the combined waves.
lam = 700e-9 # wavelength lambda
k = 2*np.pi/lam
interferEq1 = 2*I_x*(1 + np.cos(k*(2*x*a/distance)))


print('We are not able to differentiate between a left-shifted wave (i.e. left-shifted Psi)')
print('and the true wave from the left hole (i.e. PsiL) if we only plot their intensities.')
plt.plot(x, Intens, color='green', alpha = 0.5, label = '$\psi$')
plt.plot(xL, IntensR, color='yellow', alpha = 1, label = '$\psi_L$')
plt.plot(xL, Intens, color='black', linestyle = 'dashed', label = 'left-shifted $\psi$')
plt.legend()
plt.xlabel('x')
plt.ylabel('intensity')
plt.show()

print('The same thing happens for the wave from the right hole.')
plt.plot(x, Intens, color='green', alpha = 0.5, label = '$\psi$')
plt.plot(xR, IntensR, color='yellow', alpha = 1, label = '$\psi_R$')
plt.plot(xR, Intens, color='black', linestyle = 'dashed', label = 'right-shifted $\psi$')
plt.legend()
plt.xlabel('x')
plt.ylabel('intensity')
plt.show()

print('However, they are actually not the same wave! This can be demostrated when we plot their real parts instead of')
print('their intensities. We will see that they do not overlap perfectly.')
plt.plot(xL, np.real(PsiL), label = '$\psi_L$')
plt.plot(xL, np.real(Psi), label = 'left-shifted $\psi$')
# plt.plot(xR, np.real(PsiR), label = 'right')
# plt.plot(x, Intens)
# plt.plot(x, Intens)
plt.legend()
plt.xlabel('x')
plt.ylabel('intensity')
plt.show()

print('Now we plot the intensity distribution of the individual wave (PsiL and PsiR) and the combined')
print('wave (Psi1=PsiL+PsiR) to see the interference pattern visually.')
plt.plot(xL, IntensL, label = 'intensity of $\psi_L$')
plt.plot(xR, IntensR, label = 'intensity of $\psi_R$')
# plt.plot(x, Intens, label = 'middle')
plt.plot(x, IntensPsi1, label = 'interference pattern')
plt.legend()
plt.show()

print('Now we compare the result obtained by directly taking the absolute square of the combined waves and the results')
print('we obtained through the interference equation. They are purposely plotted slightly shifted from one another to')
print('make it easier to compare.')
plt.plot(x, IntensPsi1, label = 'interference pattern')
plt.plot(x+0.03, interferEq1, label = 'interference equation')
plt.legend()
plt.show()
```

### Interference pattern and phase shift (part 2) <a id='interference2'></a>

From the above cell, we see that interference pattern arises and it matches with the result we obtained through calculation using the interference equation. 

Although one can obtain the intensity distribution by directly squaring the combined amplitude $\psi_I=\psi_L+\psi_R$ using Python and obtain the interference pattern by letting Python to plot the values obtained, it is still useful to know the analytical functions sometimes, which can be obtained from the interference equation. 

Let us now compare how the intensity is distributed for:

\begin{gather}
    \Psi_{II}(\textbf{r})=\frac{1}{\sqrt{2}}\braket{\textbf{r}|L_1}-\frac{1}{\sqrt{2}}\braket{\textbf{r}|R_1}=\frac{1}{\sqrt{2}}(\Psi_L-\Psi_R)
\end{gather}

The minus sign can be incorporated into the phase of $\Psi_R$ by the relation $\exp(\pm i\pi)=-1$:

\begin{gather}
\begin{split}
    -\Psi_R(\textbf{r})&=-\frac{A_0}{d}\exp\left[-ik\left(d+\frac{x_R^2}{2d}\right)\right]\\
    &=\frac{A_0}{d}\exp(i\pi)\exp\left[-ik\left(d+\frac{x_R^2}{2d}\right)\right]\\
    &=\frac{A_0}{d}\exp\left[-i\left(k\left[d+\frac{x_R^2}{2d}\right]-\pi\right)\right]=\frac{A_0}{d}e^{i\tilde{\Phi}_R}
\end{split}
\end{gather}

And the phase difference for $\Psi_{II}$ is:

\begin{gather}
\begin{split}
    \Phi_{II}&=\tilde{\Phi}_R-\Phi_L=\left[k\left(d+\frac{x_L^2}{2d}\right)\right]-\left[k\left(d+\frac{x_R^2}{2d}\right)-\pi\right]\\
    &=k\left(\frac{x_L^2-x_R^2}{2d}\right)+\pi=-k\left(\frac{2xa}{d}\right)+\pi
\end{split}
\end{gather}

Now we can compute the intensity distribution using the interference equation (The constant prefactor is neglected for simplicity):

\begin{gather}
\begin{split}
    I_{II}(x)&=I_L+I_{\tilde{R}}+2\sqrt{I_LI_{\tilde{R}}}\cos\left[-k\left(\frac{2xa}{d}\right)+\pi\right]\\
    &=2I_0-2I_0\cos\left[k\left(\frac{2xa}{d}\right)\right]\\
    &=2I_0\left(1-\cos\left[k\left(\frac{2xa}{d}\right)\right]\right)
\end{split}
\end{gather}

where we have used: 
\begin{gather*}
    |e^{i\Phi}|=1\\
    \cos(\Phi+\pi)=\cos(\Phi)\cos(\pi)-\sin(\Phi)\sin(\pi)=-\cos(\Phi)\\
    \cos(-\Phi)=\cos(\Phi)\\
    I_L=|\Psi_L(x)|^2=I_0=I_{\tilde{R}}=|-\Psi_R(x)|^2
\end{gather*}

### Interference code cells <a id='itf_codes2'></a>

```python
Psi2 = PsiL - PsiR # Creating the combined wave Psi2

IntensPsi2 = np.absolute(Psi2)**2

# To create values for the function we obtained above using the interference equation
# which is then used to compared with the values we obtained by directly taking the absolute square of the combined waves.
lam = 700e-9 # wavelength lambda
k = 2*np.pi/lam
interferEq2 = 2*I_x*(1 - np.cos(k*(2*x*a/distance)))

print('Now we compare the result obtained by directly taking the absolute square of the combined waves and the results')
print('we obtained through the interference equation. They are purposely plotted slightly shifted from one another to')
print('make it easier to compare.')
plt.plot(x, IntensPsi2, label = 'interference pattern')
plt.plot(x+0.03, interferEq2, label = 'interference equation')
plt.legend()
plt.show()
```

Again, we see that interference pattern arises and it matches with the result we obtained through calculation using the interference equation.

Now we compute the total intensity distribution if both $\Psi_I$ and $\Psi_{II}$ are present:

\begin{gather}
\begin{split}
    I_{tot}&=I_I(x)+I_{II}(x)\\
    &=2I_0\left[\left(1+\cos\left[k\left(\frac{2xa}{d}\right)\right]\right)+\left(1-\cos\left[k\left(\frac{2xa}{d}\right)\right]\right)\right]\\
    &=2I_0\cdot 2=4I_0
\end{split}
\end{gather}


### Phase shift code cells <a id='phs_codes'></a>

```python
# The resultant intensity when both Psi1 and Psi2 are present
IntensTot = IntensPsi1 + IntensPsi2

print('Here, the 2 interference patterns are plotted separately (i.e. without taking their sum).')
print('"Interference pattern 1" is from Psi1, whereas "Interference pattern 2" is from Psi2.')
plt.plot(x, IntensPsi1, label = 'interference pattern 1')
plt.plot(x, IntensPsi2, label = 'interference pattern 2')
plt.xlabel('x')
plt.ylabel('intensity')
plt.legend()
plt.show()

print('Now we plot I_tot, which is the sum of the intensity of Psi1 and the intensity of Psi2.')
plt.plot(x, IntensTot, label = 'interference pattern (total)')
plt.xlabel('x')
plt.ylabel('intensity')
plt.legend()
plt.show()
```

### Observation <a id='observation'></a>

From the first plot, we clearly see that the two interference patterns are completely out of phase (i.e. the peaks line up perfectly with the troughs). And from the second plot, which is the sum of the two intensities, we see no interference pattern at all. 

This is because of the opposite (minus in this case) sign in $\Psi_{II}$ (which is equivalent to a $\pi$ phase shift), and causes the intensity distributions to be out-of-phase and become indecipherable.

It is uncommon to describe relativistic particles, such as photons, by assigning a wave function to it. However its success to explain the results motivates me to present it here.

Some of the reader might think that this is nothing new: number of detected photons is higher at location where the optical intensity is higher. #### First point <a id='first_pt'></a>
Going over to the second aim, which is to highlight some important points. The first point is actually mentioned earlier, which is the detection of the idler photons by the detectors $A$, $B$, $C$ or $D$, can be made to happen after their entangled pairs, the signal photons have reached the screen, by just having a longer optical path length. This is also the reason why the name of this experiment is called the delayed choice quantum eraser. 

This is mind boggling (at least to me), because it suggest retrocausality. One may argue that maybe the location of the signal photon detected on the screen is the one that dictates where its entangled pair will be detected. However, if we then replaced the beam splitters between the "which-way detectors" ($A$ and $B$) and the "information eraser" ($C$ and $D$) with fully reflective mirrors, and we can decide whether to know the path taken by the photons by reflecting the idler photons to detector $A$ and $B$, or not to know by letting the idler photons to pass through and be detected by detector $C$ and $D$, will our decision then be predetermined by the the location of the signal photons on the screen? Do we then have free will? This will certainly open up some cans of worms in philosophy.

One might think that we can then use this setup to send information backwards in time, for example a binary message by assigning '0' to no interference pattern and '1' when we observed interference pattern. However, due to the $\pi$-phase shift, we will never see interference pattern. 

But, not all hope is lost. If our explanation for the origin of the phase shift is correct, we can replace the beam splitter with other types of beam splitter that do not cause a phase shift between transmitted and reflected waves. #### Third point <a id='third_pt'></a>

Another thing that I would like to point out is that we often hear people saying that quantum objects like photons or electrons behaves like classical particles when their "path" (or the information about which slit they come from) is known. This is actually not so true. Because diffraction still happens. 

To illustrate this, we will use a normal double slits experiment. Even if we block one of the slits, we can still see diffraction pattern formed on the screen. This is not possible if they behave like classical newtonian particles. We can also see that in the calculation above.

When the "which path" information is known (here I do not mean that a conscious being is needed to know the information. The information just have to be recorded or made available at some where in the universe, for example recorded by a machine, which is not concious), the wave function only "partially collapses" (if you like Copenhagen interpretation. If you are an Everettian, then we diverge into a world line where the path is known). So, it can still undergoes diffraction.
